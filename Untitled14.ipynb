```python
import os
import nltk
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import seaborn as sns
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import calinski_harabasz_score
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import pickle
```

```python
class PretraitementLyrics:
    def __init__(self, dossier, stem=False, lemm=False):
        ##nltk.download('all')
        self.dossier = dossier
        self.stem = stem
        self.lemm = lemm

        self.stopwords_fr = stopwords.words('french')
        self.stopwords_en = stopwords.words('english')
        self.stopwords_cmr = ["a","les","parol","parole","paroles","aah","ekié","combi","diman","bro","massah","wamda","per","moto","ah","oh","eh","owoo","owooh","é","yo","refrain","couplet"]
        self.all_stopwords = list(set(self.stopwords_fr + self.stopwords_en + self.stopwords_cmr))

        self.documents = []
        self.labels = []

        self.vectorizer = None
        self.X_tfidf = None
        self.feature_names = None
        self.X_pca = None

        self.stemmer = SnowballStemmer("french")
        self.lemmatizer = WordNetLemmatizer()

    def nettoyer_texte(self, texte):
        tokens = word_tokenize(texte.lower())
        tokens = [t for t in tokens if t.isalpha() and t not in self.all_stopwords]
        if self.stem:
            tokens = [self.stemmer.stem(t) for t in tokens]
        if self.lemm:
            tokens = [self.lemmatizer.lemmatize(t) for t in tokens]
        return " ".join(tokens)

    def charger_lyrics(self):
        for fichier in os.listdir(self.dossier):
            if fichier.endswith(".txt"):
                with open(os.path.join(self.dossier, fichier), 'r', encoding='utf-8') as f:
                    texte = f.read()
                    texte_nettoye = self.nettoyer_texte(texte)
                    if texte_nettoye.strip():  # évite les textes vides
                        self.documents.append(texte_nettoye)
                        self.labels.append(fichier.split("_")[0])
        print(f"{len(self.documents)} documents chargés.")

    def vectoriser(self):
        if not self.documents:
            raise ValueError("Aucun document à vectoriser. Exécutez d'abord charger_lyrics().")
        self.vectorizer = TfidfVectorizer(stop_words=self.all_stopwords, max_features=1000)
        self.X_tfidf = self.vectorizer.fit_transform(self.documents)
        self.feature_names = self.vectorizer.get_feature_names_out()
        print("TF-IDF vectorisation terminée. Shape:", self.X_tfidf.shape)

    def afficher_wordcloud(self):
        if not self.documents:
            raise ValueError("Aucun document trouvé. Exécutez d'abord charger_lyrics().")
        texte_total = " ".join(self.documents)
        wc = WordCloud(width=800, height=400, stopwords=self.all_stopwords, background_color='white').generate(texte_total)
        plt.figure(figsize=(10, 5))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis("off")
        plt.title("WordCloud Global")
        plt.show()

    def reduire_pca(self):
        if self.X_tfidf is None:
            raise ValueError("TF-IDF non calculé. Exécutez d'abord vectoriser().")
        pca = PCA(n_components=2)
        self.X_pca = pca.fit_transform(self.X_tfidf.toarray())
        print("PCA terminé.")

    def exporter_dataframe_tfidf(self):
            if self.X_tfidf is None or self.feature_names is None:
                raise ValueError("TF-IDF non encore généré. Appelez d'abord vectoriser().")

            df = pd.DataFrame(self.X_tfidf.toarray(), columns=self.feature_names)
            df.insert(0, "Artiste", self.labels)  # Ajoute les noms d'artistes dans la 1ère colonne
            return df
    def sauvegarder_model(self, path, methode='kmeans'):
        if methode == 'kmeans':
            model = self.kmeans
        elif methode == 'gmm':
            model = self.gmm
        elif methode == 'hierarchical':
            model = self.hierarchical
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        with open(path, 'wb') as f:
            pickle.dump({
                'vectorizer': self.vectorizer,
                'model': model,
                'stopwords': self.all_stopwords,
                'feature_names': self.feature_names
            }, f)

        print(f"Modèle {methode} sauvegardé avec pickle dans : {path}")


    def charger_model(self, path, methode='kmeans'):
        with open(path, 'rb') as f:
            data = pickle.load(f)

        self.vectorizer = data['vectorizer']
        self.all_stopwords = data['stopwords']
        self.feature_names = data['feature_names']

        if methode == 'kmeans':
            self.kmeans = data['model']
        elif methode == 'gmm':
            self.gmm = data['model']
        elif methode == 'hierarchical':
            self.hierarchical = data['model']
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        print(f"Modèle {methode} chargé depuis : {path}")


    def predire_cluster(self, texte, methode='kmeans'):
        texte_nettoye = self.nettoyer_texte(texte)
        vecteur = self.vectorizer.transform([texte_nettoye])

        if methode == 'kmeans':
            cluster = self.kmeans.predict(vecteur)[0]
        elif methode == 'gmm':
            cluster = self.gmm.predict(vecteur.toarray())[0]
        elif methode == 'hierarchical':
            cluster = self.hierarchical.fit_predict(vecteur.toarray())[0]
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        print(f"Le texte est classé dans le cluster {cluster}")
        return cluster


    def afficher_documents_par_cluster(self, methode='kmeans'):
        if methode == 'kmeans':
            labels = self.kmeans_labels
        elif methode == 'gmm':
            labels = self.gmm_labels
        elif methode == 'hierarchical':
            labels = self.hierarchical_labels
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        clusters = {}
        for idx, cluster_id in enumerate(labels):
            clusters.setdefault(cluster_id, []).append(self.labels[idx])

        for cluster_id, artistes in clusters.items():
            print(f"\n--- Cluster {cluster_id} ---")
            for artiste in artistes:
                print(f"• {artiste}")
            print(f"Total documents : {len(artistes)}")
        print(f"donnees de cluster -------------------")

    def clustering_kmeans(self, n_clusters=5):
        print("\n=== Démarrage du clustering KMeans ===")
        if self.X_tfidf is None:
            raise ValueError("TF-IDF non calculé. Exécutez d'abord vectoriser().")

        print("Clustering avec KMeans...")
        self.kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        self.kmeans_labels = self.kmeans.fit_predict(self.X_tfidf)
        score = calinski_harabasz_score(self.X_tfidf.toarray(), self.kmeans_labels)
        print(f"CH Score (KMeans) = {score:.2f}")

        self.reduire_pca()  # pour affichage

        # Affichage
        plt.figure(figsize=(12, 8))
        sns.scatterplot(x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.kmeans_labels, palette='tab10')
        plt.title(f"Clustering TF-IDF avec KMeans (k={n_clusters})")
        plt.show()

        # Mots les plus représentatifs de chaque cluster
        centroids = self.kmeans.cluster_centers_.argsort()[:, ::-1]
        print("Top mots par cluster (KMeans) :")
        for i in range(n_clusters):
            top_words = [self.feature_names[ind] for ind in centroids[i, :10]]
            print(f"Cluster {i} : {', '.join(top_words)}")

    def clustering_gmm(self, n_clusters=5):
        print("\n=== Démarrage du clustering GMM ===")
        if self.X_tfidf is None:
            raise ValueError("TF-IDF non calculé. Exécutez d'abord vectoriser().")

        print("Clustering avec GMM...")
        self.gmm = GaussianMixture(n_components=n_clusters, random_state=42)
        self.gmm_labels = self.gmm.fit_predict(self.X_tfidf.toarray())
        score = calinski_harabasz_score(self.X_tfidf.toarray(), self.gmm_labels)
        print(f"CH Score (GMM) = {score:.2f}")

        self.reduire_pca()

        # Affichage
        plt.figure(figsize=(12, 8))
        sns.scatterplot(x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.gmm_labels, palette='Set2')
        plt.title(f"Clustering TF-IDF avec GMM (k={n_clusters})")
        plt.show()

    def afficher_documents_par_cluster_extrait(self, methode='kmeans', n_chars=50):
        if methode == 'kmeans':
            labels = self.kmeans_labels
        elif methode == 'gmm':
            labels = self.gmm_labels
        elif methode == 'hierarchical':
            labels = self.hierarchical_labels
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        clusters = {}
        for idx, cluster_id in enumerate(labels):
            clusters.setdefault(cluster_id, []).append((self.labels[idx], self.documents[idx]))

        for cluster_id, docs in clusters.items():
            print(f"\n--- Cluster {cluster_id} ---")
            for artiste, texte in docs:
                extrait = texte[:n_chars].replace('\n', ' ') + ("..." if len(texte) > n_chars else "")
                print(f"• {artiste} : {extrait}")
            print(f"Total documents : {len(docs)}")

    def clustering_hierarchique(self, n_clusters=5, linkage='ward'):
        if self.X_tfidf is None:
            raise ValueError("TF-IDF non calculé. Exécutez d'abord vectoriser().")

        print("Clustering avec Classification Hiérarchique Ascendante...")
        self.hierarchical = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
        self.hierarchical_labels = self.hierarchical.fit_predict(self.X_tfidf.toarray())

        score = calinski_harabasz_score(self.X_tfidf.toarray(), self.hierarchical_labels)
        print(f"CH Score (Hierarchical) = {score:.2f}")

        self.reduire_pca()

        plt.figure(figsize=(12, 8))
        sns.scatterplot(x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.hierarchical_labels, palette='Dark2')
        plt.title(f"Clustering TF-IDF avec Classification Hiérarchique (k={n_clusters})")
        plt.show()

    def dataframe_avec_clusters(self, methode='kmeans'):
        if methode == 'kmeans':
            labels = self.kmeans_labels
        elif methode == 'gmm':
            labels = self.gmm_labels
        elif methode == 'hierarchical':
            labels = self.hierarchical_labels
        else:
            raise ValueError("Méthode invalide. Choisissez 'kmeans', 'gmm' ou 'hierarchical'.")

        df = pd.DataFrame(self.X_tfidf.toarray(), columns=self.feature_names)
        df['Cluster'] = labels
        df['Artiste'] = self.labels
        return df

    def afficher_comparaison_clusters(self):
        if self.X_pca is None:
            raise ValueError("PCA non trouvée. Exécutez d'abord reduire_pca().")

        if not hasattr(self, 'kmeans_labels') or not hasattr(self, 'gmm_labels') or not hasattr(self, 'hierarchical_labels'):
            raise ValueError("Tous les clusters (kmeans, gmm, hierarchical) doivent être générés avant d'appeler cette fonction.")

        fig, axes = plt.subplots(1, 3, figsize=(20, 6))

        sns.scatterplot(ax=axes[0], x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.kmeans_labels, palette='tab10', legend=None)
        axes[0].set_title("KMeans")

        sns.scatterplot(ax=axes[1], x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.gmm_labels, palette='Set2', legend=None)
        axes[1].set_title("GMM")

        sns.scatterplot(ax=axes[2], x=self.X_pca[:, 0], y=self.X_pca[:, 1],
                        hue=self.hierarchical_labels, palette='Dark2', legend=None)
        axes[2].set_title("Hierarchical")

        for ax in axes:
            ax.set_xlabel("PCA 1")
            ax.set_ylabel("PCA 2")

        plt.suptitle("Comparaison des clusters (PCA)", fontsize=16)
        plt.tight_layout()
        plt.show()
    def afficher_pca(self):
        if self.X_pca is None:
            raise ValueError("Projection PCA non trouvée. Exécutez reduire_pca().")
        plt.figure(figsize=(12, 8))
        sns.scatterplot(x=self.X_pca[:, 0], y=self.X_pca[:, 1], hue=self.labels, palette='tab20', legend=False)
        plt.title("Projection PCA des Lyrics")
        plt.xlabel("Composante 1")
        plt.ylabel("Composante 2")
        plt.show()

```

```python
chemin = 'lyrics/lyrics_Artiste'
pt = PretraitementLyrics(dossier=chemin, stem=True, lemm=False)

pt.charger_lyrics()
pt.vectoriser()
pt.afficher_wordcloud()
pt.reduire_pca()
pt.afficher_pca()
```

```python
# Clustering KMeans
pt.clustering_kmeans(n_clusters=3)
pt.afficher_documents_par_cluster(methode='kmeans')
# Clustering GMM
pt.clustering_gmm(n_clusters=3)
pt.afficher_documents_par_cluster(methode='gmm')
# Clustering Hiérarchique Ascendant (CHA)
pt.clustering_hierarchique(n_clusters=3)
pt.afficher_documents_par_cluster(methode='hierarchical')
pt.afficher_comparaison_clusters()
```
